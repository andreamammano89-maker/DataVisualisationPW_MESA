## Airbnb Urban Pressure — Rome vs Copenhagen
### Data Visualization Project 2026

*April 2026*

In this notebook we explore Airbnb listing data for two European cities — 
Rome and Copenhagen — to understand how short-term rentals distribute 
across urban space and who controls the market.

The central question is: **how does Airbnb reshape urban neighbourhoods, 
and is the pressure concentrated or diffuse?**

We use data from [Inside Airbnb](https://insideairbnb.com/), an independent 
project that scrapes and publishes Airbnb listing data for cities worldwide. 
The Rome dataset was scraped in September 2025 (37,652 listings), 
the Copenhagen dataset in September 2025 as well (22,994 listings).

The analysis is structured in four parts:
1. Data loading and inspection
2. Cleaning and integration
3. Feature engineering (including the Urban Pressure Index)
4. Exploratory analysis to inform the visual system
