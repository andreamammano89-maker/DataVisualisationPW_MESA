In this notebook we explore Airbnb listing data for two European cities — 
Rome and Copenhagen — to understand how short-term rentals distribute 
across urban space and who controls the market.

The Rome dataset was scraped in September 2025 (37,652 listings), 
the Copenhagen dataset in September 2025 as well (22,994 listings).

The analysis is structured in four parts:
1. Data loading and inspection

4. Exploratory analysis to inform the visual system

```python
# 2. Ispezione del dataset degli annunci (Data Understanding)
print("=== ROMA: DIMENSIONE DATASET ===")
print(f"Numero di annunci: {df_rome.shape[0]}")
print(f"Numero di colonne: {df_rome.shape[1]}")

# 3. Il Piano B: Calcoliamo l'Area in km2 dei quartieri dal GeoJSON!
print("\n=== MAPPA DEI QUARTIERI E AREA ===")
# Convertiamo il sistema di coordinate a uno metrico (es. Pseudo-Mercator EPSG:3857) per misurare i metri quadri
geo_rome_metric = geo_rome.to_crs(epsg=3857)
# Calcoliamo l'area in km quadrati (diviso per 1 milione perché i risultati sono in metri quadri)
geo_rome['area_km2'] = geo_rome_metric.geometry.area / 10**6

print(geo_rome[['neighbourhood', 'area_km2']].head())
```
